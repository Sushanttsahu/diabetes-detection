```python
3import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df= pd.read_csv('diabetes.csv')
df.sample(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0.0000,np.nan)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df[df['Pregnancies']==0]
```

```python
df['Pregnancies'].unique()
```

```python
for i in df['Pregnancies'].unique():
  df[df['Pregnancies']==i]=df[df['Pregnancies']==i].fillna(df[df['Pregnancies']==i].mean().astype(int))
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.countplot(data=df,x='Outcome')
```

```python
sns.pairplot(df,hue='Pregnancies')
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
df.corr().plot(kind='kde')
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_depth=6)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=4)
```

```python
model.fit(x_train,y_train)
clf.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_predd=clf.predict(x_test)
y_pre=knn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,r2_score
score1=r2_score(y_test,y_pred)
score2=accuracy_score(y_test,y_predd)
score3=accuracy_score(y_test,y_pre)
```

```python
print(score1)
print(score2)
print(score3)
```
